```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv') # import the data, we use pandas library reads it to DataFrame
pokemon_df.head(n=5) # see the data
```

```python
### create a histogram of colum 'Speed'
plt.hist(pokemon_df['Speed'], color = 'green', edgecolor = 'black')
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.title('Histogram of Pokemon Speeds')

### Add to your histogram min, max, mean(average) lines
plt.axvline(pokemon_df["Speed"].mean(), color='grey', linestyle='dashdot', linewidth=1)
plt.axvline(pokemon_df["Speed"].max(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(pokemon_df["Speed"].min(), color='blue', linestyle='dashed', linewidth=1)
```

CORRELATION- Correlation is any statistical relationship, whether causal or not, between two random variables

```python
# SCATTER PLOT between variables: Attack and Defense
# if Attack suureneb siis ka Defense, positive correlation (aga vb olla ka negatiivne)
plt.scatter(pokemon_df['Attack'], pokemon_df['Defense'], color='blue', alpha=0.5)
plt.xlabel('Attack')
plt.ylabel('Defense')
#plt.title('Scatter Plot of Attack vs. Defense')
```

```python
# greate scattler for variables(colum): Speed and defense

plt.scatter(pokemon_df['Speed'], pokemon_df['Defense'], color='red', alpha=0.5)
plt.xlabel('Speed')
plt.ylabel('Defense')
plt.title('Scatter Plot of Speed vs. Defense')
```

```python
# scatterplot for variables: HP(health points) and Defense

plt.scatter(pokemon_df['HP'], pokemon_df['Defense'], color='purple', alpha=0.5)
plt.xlabel('HP')
plt.ylabel('Defense')
plt.title('Scatter Plot of HP vs. Defense')
```

PEARSON CORRELATION

```python
# Pearson correlation, result of this correlation calculation is between -1 and 1.
# -1 means perfect negative relationship
# 1 means perfect positive relationship
# how they inetact each other

#Weak Correlation: 0 to ±0.3
#Moderate Correlation: ±0.3 to ±0.7
#Strong Correlation: ±0.7 to ±1
```

```python
# Pearson correlation between Attack/defense

correlation1 = pokemon_df['Attack'].corr(pokemon_df['Defense'])
print(correlation1)
# moderate positive correlation, not very strong
```

```python
# Pearson correlation between Speed/Defense
correlation2 = pokemon_df['Speed'].corr(pokemon_df['Defense'])
print(correlation2)
```

```python
# Select only numeric colums

numeric_columns = pokemon_df.select_dtypes(include=['number'])
numeric_columns = numeric_columns.drop('#', axis=1)
numeric_columns
```

```python
# use for loop to....comparing target column and all other columns in the DataFrame
target_column = 'Attack'
correlation_results = {} # key = column; val= pearson_correlation
```

```python
target_column = 'Attack'
correlation_results = {}
for col in numeric_columns.columns:
    correlation = pokemon_df[target_column].corr(pokemon_df[col])
    correlation_results[col] = correlation

correlation_results
```

```python
# Select only numeric columns

numeric_columns = pokemon_df.select_dtypes(include=['number'])
numeric_columns = numeric_columns.drop('#', axis=1)
numeric_columns
target_column = 'Attack'
correlation_results = {} #key = column, val=pearson_correlation

# for loop comparing target column and all other columns in the DataFrame

for col in numeric_columns.columns:
  if col =='Attack':
    continue
  correlation = pokemon_df[target_column].corr(pokemon_df[col])
  correlation_results[col] = correlation

correlation_results
```

CORRELATION MATRIX

```python
# IS MATRIX, that

correlation_matrix = numeric_columns.corr() # all columns are numeric. we can do the same with doble for loops
correlation_matrix
```

LINEARS REGRESSION- predict one value based on another

```python
# linear regression- predict one value based on another;  IN ONE of the simplest machine learning..
```

```python
# regression plots, ading linear regression line

sns.regplot(x = 'Attack', y = 'Defense', data = pokemon_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Create two DataFrame Grass and Water, filter out grass and water pokemons
grass_df = pokemon_df[pokemon_df['Type 1'] == 'Grass']
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']

```

```python
#print(grass_df.head(n=10))
```

```python
#Create the regression plots for each (Grass and Water)
# Grass
sns.regplot(x='Attack', y='Defense', data=grass_df, color='green', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Grass Type pokemon, Attack vs Defense')
plt.show()
```

```python
#Create the regression plots for each (Grass and Water)
# Water
sns.regplot(x='Attack', y='Defense', data=water_df, color='blue', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Water Type pokemon, Attack vs Defense')
plt.show()
```

```python
#Calculate the Pearson correlation for each DataFrame (variables: Attack and Defense)

correlation3 = grass_df['Attack'].corr(grass_df['Defense'])
print('Pearson correlation for Grass pokemons:', correlation3)

correlation4 = water_df['Attack'].corr(water_df['Defense'])
print('Pearson correlation for Water pokemons:', correlation4)
```
